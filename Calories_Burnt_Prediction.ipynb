Tmporting the Dependences

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection & Prediction

```python
from google.colab import files
uploaded = files.upload()

calories = pd.read_csv("calories.csv")
exercise = pd.read_csv("exercise.csv")


```

```python
exercise.head()
```

Combining the two Dataframes

```python
calories_data = pd.concat([exercise, calories['Calories']], axis=1)
calories_data.head()
```

```python
#cheaking the number of rows and columns
calories_data.shape
```

```python
#Getting some information about the data
calories_data.info()
```

Converting the text data to numerical values

```python
calories_data['Gender'] = calories_data['Gender'].map({'male': 0, 'female': 1})
```

Finding the correlation in the Dataset

1.Positive Correlation

2.Negative Correlation

```python
correlation = calories_data.corr()
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Separating features and Target

```python
x = calories_data.drop(columns=['User_ID', 'Calories'], axis=1)
y = calories_data['Calories']
```

```python
print(x)
```

```python
print(y)
```

Sprinting the data into Training data and Test Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Model Training

```python
#Loading the model
model = XGBRegressor()
```

```python
#training the model with x_train
model.fit(x_train, y_train)
```

Evaluation

Prediction on test Data

```python
test_data_prediction = model.predict(x_test)
```

```python
print(test_data_prediction)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(y_test, test_data_prediction, c='crimson')

# Plotting a line for perfect prediction
p1 = max(max(test_data_prediction), max(y_test))
p2 = min(min(test_data_prediction), min(y_test))
plt.plot([p1, p2], [p1, p2], 'b-')

plt.xlabel('Actual Calories', fontsize=15)
plt.ylabel('Predicted Calories', fontsize=15)
plt.title('Actual vs Predicted Calories')
plt.show()
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(y_test, test_data_prediction)
print("The Mean Absolute Error: ", mae)
```
