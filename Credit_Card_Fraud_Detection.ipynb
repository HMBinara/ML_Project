```python
#seperating the data for analysis
legit = creadit_card_data[creadit_card_data.Class == 0]
fraud = creadit_card_data[creadit_card_data.Class == 1]
```

#Under-Sampling

Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction

number of fraudulent transactin -->492


```python
legit_sample = legit.sample(n = 492)
```


Concatenating two DataFrames
---



```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Target

```python
x= new_dataset.drop(columns='Class',axis=1)
y= new_dataset['Class']
```

Split the data into Training data &Testing Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

#Model Traning

```python
model = LogisticRegression()
```

```python
#training the logistic Regression Model with Training Data
model.fit(x_train,y_train)
```

Evaluation

```python
from sklearn.metrics import accuracy_score

x_train_prediction = model.predict(x_train)
training_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print(training_accuracy)
```

```python
#accurcy on test data
x_test_prediction = model.predict(x_test)
test_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print(test_accuracy)
```
